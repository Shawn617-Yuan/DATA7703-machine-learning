```python
from datetime import date
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns
df=pd.read_csv("/Users/jiezhang/Downloads/vscode/machine learning/practise/dataset/pokemonfun.csv")
data=df.dropna()
```

```python
from sklearn.preprocessing import StandardScaler
X=np.array(data.iloc[:,:8])
# Standardizing the features
X1 = StandardScaler().fit_transform(X)
```

```python
pca=PCA(n_components=2)
pca.fit_transform(X1)
# eigenvector
print(pca.components_)
```

```python
# eigenvalue
print("eigenvalue:" f"{pca.explained_variance_}")
```

```python
# percentage of variables one pcx can represent 
print(pca.explained_variance_ratio_)
```

```python
# reduced to 2 dimensions: PC1 and PC2
pc=pca.fit_transform(X1)
print(pc)
```

```python
PC_values = np.arange(pca.n_components_) + 1
plt.plot(PC_values, pca.explained_variance_ratio_, 'go-', linewidth=3)
plt.xlabel("PC values")
plt.ylabel("percentage of variables represent")
plt.show()
```

```python
plt.scatter(pc[:,0], pc[:,1])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```
